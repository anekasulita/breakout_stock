```python
import pandas as pd
```

```python
import os
list_file = os.listdir("/Users/asulita1/downloads/saham/")
```

```python
full_df = pd.DataFrame()
for each in list_file:
    str1 = "/Users/asulita1/downloads/saham/"+each
    day_df = pd.read_csv(str1)
    full_df = full_df.append(day_df,ignore_index = True)
```

```python
list_ticker = full_df['<ticker>'].unique()
print(type(list_ticker))
```

```python

from datetime import date
import datetime
 
today = date.today() - datetime.timedelta(days=1)
today_str = today.strftime("%m/%d/%y")

print("Today's date:", today_str)
```

```python
summary_df = pd.DataFrame(columns=['<ticker>','<close>','<max_high>','<break_high>','<volume>','<avg_volume>','<double_volume>'])

for each in list_ticker:
    ticker = each
    close_bool = ((full_df.loc[:,"<ticker>"]==each)&(full_df.loc[:,"<date>"]==today_str))
    
    close_list = full_df.loc[close_bool,"<close>"]
    if close_list.empty:
        close = 0
    else:
        close = close_list.iloc[0]
    max_high = full_df.loc[full_df.loc[:,"<ticker>"]==each,"<close>"].max()
    break_high = close == max_high
    volume_bool = ((full_df.loc[:,"<ticker>"]==each)&(full_df.loc[:,"<date>"]==today_str))
    volume_list = full_df.loc[volume_bool,"<volume>"]
    if volume_list.empty:
        volume = 0
    else:
        volume = volume_list.iloc[0]
    avg_volume = full_df.loc[full_df.loc[:,"<ticker>"]==each,"<volume>"].mean()
    double_volume = volume > (avg_volume * 1.5)
    new_record = pd.DataFrame([[ticker, close, max_high, break_high, volume, avg_volume, double_volume]],columns=['<ticker>','<close>','<max_high>','<break_high>','<volume>','<avg_volume>','<double_volume>'])
    summary_df = summary_df.append(new_record, ignore_index = True)
    
    #print("each : -",each,"- close: -",close,"- max_high: -",max_high,"- break_high: -",break_high,"- volume: -",volume,"- avg_volume: -",avg_volume,"- double_volume : -",double_volume,"-")
    
```
